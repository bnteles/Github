```python
import pandas as pd
```

```python
df2017 = pd.read_csv("2017 BEATRIZ TELES - VIOLENCIA POLICIAL.csv")
```

```python
df2017["TEMA_ESPECIFICO"].unique()
```

```python
df2017["TEMA_ESPECIFICO"] = df2017["TEMA_ESPECIFICO"].replace("Violencia policial/Racismo","Violencia policial / Racismo")
```

```python
df2017 = df2017.drop(columns=["Resumo (TRANSCRIÇÃO)"])
```
